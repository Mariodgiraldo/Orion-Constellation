```python
#1 
#Importing
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#2
#each constellation coordinate seems to fall within a certain range
## x = -0.5 : 0.57
## y = 2 : 14
## z = 0 : 3

x = [-0.41, 0.57, 0.07, 0.00, -0.29, -0.32,-0.50,-0.23, -0.23]
y = [4.12, 7.71, 2.36, 9.10, 13.35, 8.13, 7.19, 13.25,13.43]
z = [2.06, 0.84, 1.56, 2.07, 2.36, 1.72, 0.66, 1.25,1.38]

#3 
#2D visualization

orion = plt.figure(figsize=(10,5))

ax1 = orion.add_subplot(1,2,1,facecolor = 'black')
plt.scatter(x,y)
#ax1.set_size_inches(10,8)
plt.title('2D Visualization of Orion')
plt.show
plt.clf

#4
#3D visualization
ax2 = orion.add_subplot(1,2,2, projection='3d', facecolor = 'black')
plt.scatter(x,y,z, c = 'gold', marker = '*')
#ax1.set_size_inches(10,8)
plt.title('3D Visualization of Orion')
plt.show
plt.clf

#5 
#Adjusting background of plots to look like night sky
#adding code to #4
```
